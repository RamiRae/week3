```python
import pandas as pd
import statsmodels.formula.api as sm
```

```python
data = pd.read_csv('MachineLearningRating_v3.csv')  # Replace 'your_data.csv' with your actual file name

# Assuming your columns are named 'Province' and 'TotalClaims'
# Adjust column names if necessary
model = sm.ols('TotalClaims ~ Province', data=data)
results = model.fit()

# Print the ANOVA table
print(results.summary())

# Access the F-statistic and p-value
f_statistic = results.fvalue
p_value = results.f_pvalue

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is evidence of a difference in risk across provinces.")
else:
    print("Fail to reject the null hypothesis. There is no evidence of a difference in risk across provinces.")
```

```python
model = sm.ols('TotalClaims ~ PostalCode', data=data)
results = model.fit()

# Print the ANOVA table
print(results.summary())

# Access the F-statistic and p-value
f_statistic = results.fvalue
p_value = results.f_pvalue

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is evidence of a difference in risk between zip codes.")
else:
    print("Fail to reject the null hypothesis. There is no evidence of a difference in risk between zip codes.")
```

```python
data['ProfitabilityIndex'] = data['TotalPremium'] - (data['TotalClaims'] / data['SumInsured'])

# Create the model
model = sm.ols('ProfitabilityIndex ~ PostalCode', data=data)
results = model.fit()

# Print the ANOVA table
print(results.summary())

# Access the F-statistic and p-value
f_statistic = results.fvalue
p_value = results.f_pvalue

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is evidence of a significant difference in profitability between zip codes.")
else:
    print("Fail to reject the null hypothesis. There is no evidence of a significant difference in profitability between zip codes.")
```
